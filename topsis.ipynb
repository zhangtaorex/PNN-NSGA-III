```python
import pandas as pd
import numpy  as np
import math
```

```python
objs = pd.read_csv("./proto_resultfinal.csv",header=None)
objs = objs.drop(columns=[0,2])
objs = objs.drop(index=0)
objs = objs.drop_duplicates()
# objs.to_csv("./objfinal.csv")
obj = objs.to_numpy(dtype=np.float64)
```

```python
max_mass = max(obj[:,0])
min_mass = min(obj[:,0])
max_load = max(obj[:,1])
min_load = min(obj[:,1])
obj[:,0] = (obj[:,0]-min_mass)/(max_mass-min_mass)
obj[:,1] = (obj[:,1]-min_load)/(max_load-min_load)
```

```python
obj[obj[:,0] ==0] = np.array([1e-8,1e-8]) 
obj
```

```python
q = np.zeros((obj.shape))
for i in range(obj.shape[1]):
    sum = np.sum(obj[:,i])
    for j in range(obj.shape[0]):
        q[j,i] = obj[j,i]/(sum-obj[j,i])
```

```python
l = np.zeros(2)
l[0] = np.sum(q[:,0]*np.log(q[:,0]))
l[1] = np.sum(q[:,1]*np.log(q[:,1]))
```

```python
l = -l/np.log(2)
```

```python
w = np.zeros(2)
```

```python
w[0] = (1-l[0])/(2-l[0]-l[1])
```

```python
w[1] = (1-l[1])/(2-l[0]-l[1])
```

```python
w
```

```python
h = np.zeros((obj.shape))
```

```python
for i in range(h.shape[1]):
    for j in range(h.shape[0]):
        h[j,i] = q[j,i]*w[i]
```

```python
a_plus = (np.min(h[:,0]), np.max(h[0,:]))
a_mus = (np.max(h[:,0]), np.min(h[0,:]))
a_plus
```

```python
a_mus
```

```python
d_plus = np.zeros(obj.shape[0])
d_mus = np.zeros(obj.shape[0])
```

```python
for i in range(len(d_plus)):
    d_plus[i] = math.sqrt((a_plus[0]-h[i,0])**2+(a_plus[1]-h[i,1])**2)
    d_mus[i] = math.sqrt((a_mus[0]-h[i,0])**2+(a_mus[1]-h[i,1])**2)
```

```python
T = np.zeros(obj.shape[0])
T = d_mus/(d_plus+d_mus)
```

```python
T_MAX = max(T)
T_MIN = min(T)
```

```python
t = (T-T_MIN)/(T_MAX-T_MIN)
t =  pd.DataFrame(t)
t.to_csv("./t_final.csv")
```
